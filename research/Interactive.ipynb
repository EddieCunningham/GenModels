```python
import networkx as nx

from bokeh.plotting import figure
from bokeh.io import show, output_notebook
from bokeh.models import Plot, Range1d, MultiLine, Circle, HoverTool, TapTool, BoxSelectTool
from bokeh.models.graphs import from_networkx, NodesAndLinkedEdges, EdgesAndLinkedNodes
from bokeh.palettes import Spectral4
from functools import partial
```

```python
G = nx.DiGraph()
G.add_edge( 1, 2, data='0' )
G.add_edge( 3, 2, data='1' )
G.add_edge( 2, 5, data='2' )
G.add_edge( 5, 7, data='3' )
```

```python
for a, b, c in G.edges.data( 'data' ):
    print( a, b, c )
```

```python
layout = nx.nx_agraph.pygraphviz_layout( G, prog='dot' )
positions = np.array( list( layout.values() ) )
x_min, x_max = positions[ :, 0 ].min() - 10, positions[ :, 0 ].max() + 10
y_min, y_max = positions[ :, 1 ].min() - 10, positions[ :, 1 ].max() + 10
```

```python
graph_renderer = from_networkx( G, partial( nx.nx_agraph.pygraphviz_layout, prog='dot' ) )

graph_renderer.node_renderer.glyph = Circle(size=15, fill_color=Spectral4[0])
graph_renderer.node_renderer.selection_glyph = Circle(size=15, fill_color=Spectral4[2])
graph_renderer.node_renderer.hover_glyph = Circle(size=15, fill_color=Spectral4[1])

graph_renderer.edge_renderer.glyph = MultiLine(line_color="#CCCCCC", line_alpha=0.8, line_width=5)
graph_renderer.edge_renderer.selection_glyph = MultiLine(line_color=Spectral4[2], line_width=5)
graph_renderer.edge_renderer.hover_glyph = MultiLine(line_color=Spectral4[1], line_width=5)

graph_renderer.selection_policy = NodesAndLinkedEdges()
graph_renderer.inspection_policy = EdgesAndLinkedNodes()

x_range = Range1d( x_min, x_max )
y_range = Range1d( y_min, y_max )
fig = figure( tools='hover,crosshair,pan,wheel_zoom,reset,tap,box_select', x_range=x_range, y_range=y_range )

fig.renderers.append( graph_renderer )

output_notebook()
show( fig )
```
